```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
#file path in Google Drive
file_path = "/content/drive/My Drive/BeforePythonGoogleApps34.xlsx"
# Load the Excel file
dataGoogle = pd.read_excel(file_path)

# Display the first few rows to confirm it loads correctly
dataGoogle.head()

```

```python
# Display dataset info (number of rows and columns)
print("\nDataset Info:")
print(dataGoogle.info())

# Display number of rows and columns
print("\nNumber of Rows:", dataGoogle.shape[0])
print("Number of Columns:", dataGoogle.shape[1])
```

```python
#1 Filtering the data based on Rating.

DataGoogleRating = dataGoogle[dataGoogle['score'] >= 2.5]

# Display dataset info (number of rows and columns)
print("\nDataset Info:")
print(DataGoogleRating.info())

```

```python
GoogleDATE1 = DataGoogleRating
print(GoogleDATE1.info())
```

```python
# Convert the nanosecond timestamps to datetime format
GoogleDATE1["updated"] = pd.to_datetime(GoogleDATE1["updated"], unit='s')
```

```python
# Format the date into 'YYYY-MM-DD' format
GoogleDATE1["updated"] = GoogleDATE1["updated"].dt.strftime("%Y-%m-%d")

# Display the updated column
print("\nðŸŽ¯ Final formatted dates:\n", GoogleDATE1[["appId", "updated"]].head(10))
```

```python
Google5yrDATE1= GoogleDATE1
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Define the date 5 years ago from today
five_years_ago = datetime.today() - timedelta(days=5*365)

# Ensure 'updated' column is in datetime format
Google5yrDATE1 = Google5yrDATE1.copy()  # Avoid SettingWithCopyWarning
Google5yrDATE1.loc[:, "updated"] = pd.to_datetime(Google5yrDATE1["updated"], format="%Y-%m-%d", errors='coerce')

# Filter the dataset to keep only apps updated in the last 5 years
recent_apps_5yrs = Google5yrDATE1[Google5yrDATE1["updated"] >= five_years_ago]

# Display first few rows
print(recent_apps_5yrs.head())


print(recent_apps_5yrs.info())
```

```python
# Define the output file path in Google Drive
output_file_path = "/content/drive/My Drive/PostPythonGoogle.xlsx"

# Save the DataFrame as an Excel file
recent_apps_5yrs.to_excel(output_file_path, index=False)  # index=False prevents adding an extra index column

print(f"File saved successfully at: {output_file_path}")
```
